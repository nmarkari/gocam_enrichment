**Note on C4B:** 

C4B is listed twice, but that is due to  P0C0L4 and P0C0L5 both being officially listed as uniprot accession IDs for C4B, which affected our results when we used SimpleMine. 

P0C0L4 seems to actually correspond to C4A (C4A and C4B are paralogs). However, C4b can be derived from either the C4A gene product or the C4B gene product, and so both genes are listed as members of a set, and both are annotated to the pathway. GO-CAMs use uniprot accession IDs, so P0C0L4 and P0C0L5 are listed in the model for enrichment. When the input list of genes was mapped from gene symbols to uniprot accession IDs, both P0C0L4 and P0C0L5 were hits for the GOCAM. 

Why both accession IDs are officially associated with C4B is unclear, since they correspond to C4A and C4B on UniProt's website. We believe the issue is that the genes are cross listed as aliases for one another, which led to both uniprot accession ID's being associated with C4B.
